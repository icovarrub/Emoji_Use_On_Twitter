```python
from nltk.tokenize.casual import TweetTokenizer
tokenizer = TweetTokenizer(strip_handles=True, preserve_case=False)
```

```python
tokenizer.tokenize(tweets['full_text'].loc[0])
```

```python
def process_tweet(t):
    emojis = set(demoji.findall_list(t, desc=False))
    tokens = tokenizer.tokenize(demoji.replace(t, ' '))
    return {'emojis':emojis, 'tokens':tokens}
```

```python
process_tweet(tweets['full_text'].loc[49])
```

```python
tokenized = pd.DataFrame(map(process_tweet, tqdm(tweets['full_text'])))
```

```python
import multiprocessing as mp
```

```python
with mp.Pool() as p:
    tokenized = pd.DataFrame(p.imap(process_tweet, tqdm(tweets['full_text']), chunksize=1000))
```

```python
tokenized.head(50)
```

```python
from collections import Counter
```

```python
def count(tokens):
    return Counter(concat(tokens))
```

```python
tok_freq = count(tokenized['tokens'])
```

```python
tok_freq.most_common(10)
```

```python
emojis = count(tokenized['emojis'])
```

```python
emojis.most_common(10)
```

```python
exploded = tokenized.explode('emojis').dropna()
exploded.head(10)
```

```python
N = sum(tok_freq.values())
```

```python
N
```

```python
from math import log2

def pmi(f12, f1, f2, N):
    return log2(f12) + log2(N) - log2(f1) - log2(f2)
```

```python
import nltk
stop = nltk.corpus.stopwords.words('english')
```

```python
import re
```

```python
terms = Counter()
emoji_tok_freq = count(exploded[exploded['emojis']=='🤖']['tokens'])
emoji_N = sum(emoji_tok_freq.values())
for t in emoji_tok_freq:
    if t not in stop and emoji_tok_freq[t] > 20 and re.search(r'[a-z]', t):
        terms[t] = pmi(emoji_tok_freq[t], tok_freq[t], emoji_N, N)
```

```python
terms.most_common(10)
```

Compute PMI word associations for all emojis

```python
for e in emojis:
    if emojis[e] > 50:
        terms = Counter()
        emoji_tok_freq = count(exploded[exploded['emojis']==e]['tokens'])
        emoji_N = sum(emoji_tok_freq.values())
        for t in emoji_tok_freq:
            if t not in stop and emoji_tok_freq[t] > 20 and re.search(r'[a-z]', t):
                terms[t] = pmi(emoji_tok_freq[t], tok_freq[t], emoji_N, N)
        print(e, ', '.join(map(first, terms.most_common(10))))
```

### Final product: a table showing words that are associated (by G2 or some other metric) with each emoji in the corpus

```python
import llr
from math import log2
```

```python
def G2(f12, f1, f2, N):
    return llr.llr_root(f12, f1-f12, f2-f12, N-f1-f2+f12)
```

```python
for e in emojis:
    if emojis[e] > 50:
        terms = Counter()
        emoji_tok_freq = count(exploded[exploded['emojis']==e]['tokens'])
        emoji_N = sum(emoji_tok_freq.values())
        for t in emoji_tok_freq:
            if t not in stop and emoji_tok_freq[t] > 20 and re.search(r'[a-z]', t):
                terms[t] = G2(emoji_tok_freq[t], tok_freq[t], emoji_N, N)
        print(e, ', '.join(map(first, terms.most_common(10))))
```
